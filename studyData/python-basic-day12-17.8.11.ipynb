```python
import threading
import time
```

```python
# 전역 변수
# 목적. thread를 여러개 돌려서 race condition 을 의도적으로 일으켜서 확인함.
# 5,000,000 -> thread 를 50개를 100,000 개씩 전역 변수를 올리게함.
g_count = 0
lock = threading.Lock()
```

```python
def thread_main():
    global g_count
    lock.acquire()
    # 함수 내에서 접근해서 변경하기 위해서는 global 을 사용해줌. 
    
    for i in range(100000):
        g_count += 1
    
    lock.release()
        
```

```python
threads = []
```

```python
start_time = time.time()
for i in range(50):
    th = threading.Thread(target = thread_main)
    threads.append(th)
    # thread 50개 만듬.
    
for th in threads:
    th.start()
    
for th in threads:
    th.join()
    
print("g_count : {:,}".format(g_count))
end_time = time.time()

print(end_time-start_time)

```
