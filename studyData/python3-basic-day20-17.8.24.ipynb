```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com/"
html = urllib.request.urlopen(url) # urllib으로 open 하는곳이 

# print(html) 현재 response에 남아 있는것
source = html.read()
html.close()

# print(source)
# 여기까지가 html 코드를 받아온것. 

soup = BeautifulSoup(source, "html.parser")
# print(soup) # html 의 코드를 받은것.

table = soup.find(id="Top-Box-Office") # find 라는 내장 함수를 이용해서 id가 Top-Box-Office 인 테그 값을 찾아낼수있음

print(table)
```

```python
all_tr = table.find_all("tr") # find_all 을하면 가장 위에서 한개씩 모든것을 리스트로 묶음
print(all_tr)
```

```python
for tr in all_tr:
    all_td = tr.find_all("td") # 해당 raw 안에 해당 데이터.. 
    score = all_td[0].find("span", attrs={"class":"tMeterScore"}).text # 조건을 줄수 있음. , attrs={"":""}
    movie_name = all_td[1].a.text
    amount = all_td[2].a.text.strip() # amunt는 all_td 의 2번째의 a 테그의 text 
    
    # .strip() 를 써주면, 스페이스가 되어있는 값들을 한줄로 묶어주는 명령어
    print("0."+score, movie_name, amount, "\n") # score 는 문자열이기 떄문에, "0.+" 가 연산이 가능함.
    
```

```python
import os
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.by import By
 
```

```python
ff_driver = webdriver.Chrome()#Firefox()
ff_driver.get("https://www.google.co.kr/")

# 웹드라이브
```

```python
query = ff_driver.find_element_by_id("lst-ib") #크롬에서 검색값 id = lst-ib 
 
query.send_keys("linus torvalds nvidia") # 검색창에 적는것. 


# 파이어폭스가 크롬보다 안전함.. 오 ..

ff_driver.find_element_by_name("btnK").click() # 이것은 검색 버튼을 누르게 하는것.
ff_driver.implicitly_wait(10) # 인터넷 속도 환경에 따라서 다음 액션 사이의 발생하는 시간이 ㅇ 있을경우 그것을 막아주기 위해.


RESULTS_LOCATOR = "//div/h3/a" # 앞의 가장 공통되는 부분을 선택하는것인데...?


WebDriverWait(ff_driver, 10).until(EC.visibility_of_element_located((By.XPATH, RESULTS_LOCATOR))) 
 

#results = ff_driver.find_elements_by_xpath(RESULTS_LOCATIR)
results = ff_driver.find_elements(By.XPATH, RESULTS_LOCATOR) 


for item in results:
    print(item.text)
```
